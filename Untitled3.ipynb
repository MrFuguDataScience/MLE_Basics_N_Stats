# $\hat{f}$= **aX + B**

The $\hat{f}$ represents that we are `estimating` or you can say `"predicting"` the data from our function

+ `a` is the slope (coefficient)
+ `b` intercept
+ `X` is our response variable "independent"

# `More terminology that you will see but, don't get confused:`

+ The **`response variable`** is the **`dependent variable`** that we are trying to **`explain`** AKA *`Y`* because it depends on another variable.

+ The **`explanatory variable`**, is used to **`predict/explain`** the response variable. This is your **`Independent`** variable *`X`*

`---------------------------------------------------------------------`

If we were to expand this in terms of simple `linear regression` we would note that if `two variables` have a `strong relationship` with each other, this will describe a line.

The best fit for this line with help us make predictions based on our data. 

**`Two things to think about here and distinguish`**

1.) For example, if we have a given pair of points that exists and we want take a vertical arrow line down to the line that runs near this point(s) we are given the predicted values. The distance (vertical) from each point to that line will have error. The error for all points needs to be minimized meaning that the distance from all points will adjusted in order to make an overall error as small as possible for each point as a cummulative value.

2.) The actual data point we are given can be used to find predicted values that are unknown. These values will have some inherent error and will more than likely not actually exist on the line as observed values. Interestingly, you can view this another way where you have an actual value as your data point and the predicted value on the line vertically down from that point placed on the line. The distance from your point to that line is there error.

+ To calculate this we use `Least Squares` we will get the squared distance between the point and our line for each point to minimize the total error for our dataset.

+ We would say that the data point we observe is ( x, y ) and the predicted on the line vertically down from the point onto the line is ( x, $\hat{y}$ )

This is the difference between the `observed` value $y_i$ which is a data point you have in the dataset and the `predicted value` from that particular observation $\hat{f}(x_i)$.

+ Evaluating a `goodness-of-fit` we are looking for `smaller values` of total error!

`------------------------------------------------------------------`
